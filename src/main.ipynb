## Elementos de un Diagrama de Caja


  <td align="center" style="padding: 0 20px;">
        <img src="../img/image4.png" alt="" width="400">
      </td>

Un diagrama de caja (o *boxplot*) resume un conjunto de datos numéricos mediante cinco valores principales y posibles valores atípicos. A continuación, se describen sus elementos:

### 1. **Caja (o rectángulo)**
- **Límite inferior de la caja:** Representa el primer cuartil (**Q1**), que delimita el 25% inferior de los datos.
- **Límite superior de la caja:** Representa el tercer cuartil (**Q3**), que delimita el 75% inferior de los datos (o el 25% superior).
- **Línea dentro de la caja:** Muestra la mediana (**Q2**), que es el valor central del conjunto de datos.

### 2. **Bigotes**
- Los bigotes son líneas que se extienden desde los límites de la caja hasta los valores más extremos que no se consideran atípicos.
- Se calculan generalmente como:
  - **Límite inferior:** $Q1 - 1.5 \times \text{IQR}$
  - **Límite superior:** $Q3 + 1.5 \times \text{IQR}$
  - Donde $ \text{IQR} $ (rango intercuartílico) es $ Q3 - Q1 $.

### 3. **Valores atípicos (outliers)**
- Son puntos individuales fuera de los límites definidos por los bigotes.
- Se representan con símbolos como círculos o asteriscos.

### 4. **Rango intercuartílico (IQR)**
- Aunque no se dibuja directamente, es una medida importante para calcular los bigotes:
  $$
  \text{IQR} = Q3 - Q1
  $$

### 5. **Escala numérica**
- El eje del diagrama, generalmente el eje x o y, muestra los valores del conjunto de datos.

Este diagrama es útil para identificar la **dispersión**, **simetría** y **valores atípicos** de los datos.

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['ingreso_total'], color='lightgreen')
plt.title('Boxplot de Ingresos Totales')
plt.xlabel('Ingreso Total')
plt.show()
```

El boxplot que muestras parece indicar que los datos de ingreso total están altamente concentrados en valores bajos, mientras que hay un número significativo de valores atípicos (outliers) en rangos altos. Esto es común en distribuciones de ingresos, ya que la mayoría de las personas tienen ingresos dentro de un rango más estrecho, y unos pocos casos extremos (como altos ejecutivos o empresarios) tienen ingresos muy elevados.

## Mejora del Análisis del Gráfico

Para entender mejor la distribución:

```python
# Calcular estadísticas descriptivas
Q1 = df['ingreso_total'].quantile(0.25)  # Primer cuartil
Q2 = df['ingreso_total'].median()        # Mediana
Q3 = df['ingreso_total'].quantile(0.75)  # Tercer cuartil
IQR = Q3 - Q1                                 # Rango intercuartílico
upper_bound = Q3 + 1.5 * IQR                  # Límite superior
lower_bound = Q1 - 1.5 * IQR                  # Límite inferior

# Valores superiores al límite superior
outliers = df[df['ingreso_total'] > upper_bound]
num_outliers = len(outliers)

# Mostrar estadísticas
print("Primer cuartil (Q1):", Q1)
print("Mediana (Q2):", Q2)
print("Tercer cuartil (Q3):", Q3)
print("Rango intercuartílico (IQR):", IQR)
print("Límite superior:", upper_bound)
print("Número de valores superiores al límite superior:", num_outliers)

# Crear el boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['ingreso_total'], color='lightgreen')
plt.title('Boxplot de Ingresos Totales con Límites')
plt.xlabel('Ingreso Total')

# Añadir líneas para los cuartiles y la mediana
plt.axvline(Q1, color='blue', linestyle='--', label='Q1 (25%)')
plt.axvline(Q2, color='red', linestyle='-', label='Mediana (50%)')
plt.axvline(Q3, color='blue', linestyle='--', label='Q3 (75%)')

# Añadir límites al gráfico
plt.axvline(upper_bound, color='orange', linestyle='--', label='Límite Superior')
plt.axvline(lower_bound, color='orange', linestyle='--', label='Límite Inferior')

# Mostrar la leyenda
plt.legend()
plt.show()
```

```python
# Valores dentro de la caja (entre Q1 y Q3)
within_box = df[(df['ingreso_total'] >= Q1) & (df['ingreso_total'] <= Q3)]
num_within_box = len(within_box)

# Valores fuera de los bigotes
outliers = df[(df['ingreso_total'] < lower_bound) | (df['ingreso_total'] > upper_bound)]
num_outliers = len(outliers)

# Total de datos
total_data = len(df)

# Porcentajes
percent_within_box = (num_within_box / total_data) * 100
percent_outliers = (num_outliers / total_data) * 100

# Mostrar resultados
print("Porcentaje de datos dentro de la caja (Q1-Q3): {:.2f}%".format(percent_within_box))
print("Porcentaje de datos fuera de los bigotes: {:.2f}%".format(percent_outliers))
```

```python
# Definir el salario mínimo
salario_minimo = 156000

# Filtrar los datos
superan_salario_minimo = df[df['ingreso_total'] > salario_minimo]
no_superan_salario_minimo = df[df['ingreso_total'] <= salario_minimo]

# Contar los datos
num_superan = len(superan_salario_minimo)
num_no_superan = len(no_superan_salario_minimo)

# Total de datos
total_datos = len(df)

# Calcular porcentajes
percent_superan = (num_superan / total_datos) * 100
percent_no_superan = (num_no_superan / total_datos) * 100

# Mostrar resultados
print("Número de ingresos que superan el salario mínimo:", num_superan)
print("Porcentaje de ingresos que superan el salario mínimo: {:.2f}%".format(percent_superan))
print("Número de ingresos que no superan el salario mínimo:", num_no_superan)
print("Porcentaje de ingresos que no superan el salario mínimo: {:.2f}%".format(percent_no_superan))
```

De un total de **16,846 ingresos analizados**, se observa lo siguiente:

1. **El 62.69% de los ingresos (10,561 casos)** superan el salario mínimo de **156,000 pesos**.  
   Esto indica que la mayoría de las personas tienen ingresos por encima de este umbral, lo que podría reflejar una distribución de ingresos relativamente favorable para este grupo.

2. **El 37.31% de los ingresos (6,285 casos)** no alcanzan el salario mínimo.  
   Esto representa a una proporción significativa de la población que podría estar en una situación de vulnerabilidad económica, dependiendo de factores como el costo de vida en su región o sus responsabilidades financieras.

# Datos del Dataset que Podrían No Ser de Interés

Para determinar qué datos podrían no ser de interés, evaluamos si cumplen con alguna de las siguientes características:

1. **Redundancia:** Datos repetidos o que ya están representados en otras columnas.
2. **Relevancia:** Datos que no aportan valor al análisis actual o que no están relacionados con los objetivos del estudio.
3. **Falta de variabilidad:** Columnas donde todos los valores (o casi todos) son iguales, lo que no permite identificar diferencias significativas.

---

## Posibles datos no relevantes para ciertos análisis

1. **`peso`:**  
   - Representa el factor de ponderación de cada observación. Si no se está realizando un análisis representativo a nivel poblacional, podría no ser útil.

2. **`habilidad_lectura_escritura`:**  
   - Tiene poco valor en un análisis económico, ya que la mayoría de los individuos empleados probablemente tengan esta habilidad, y podría no mostrar variabilidad significativa.

3. **`asistencia_educativa` (`CH10`):**  
   - Indica si el individuo asiste a un establecimiento educativo. Esto podría ser irrelevante si el análisis está enfocado en ingresos, empleo o región.

4. **`tipo_institucion_educativa` (`CH11`):**  
   - Si el interés principal está en los ingresos o el empleo, este dato podría no ser relevante, ya que se relaciona más con la educación de niños o jóvenes.

5. **`nacionalidad`:**  
   - Podría ser irrelevante si no se planea investigar diferencias en ingresos o empleo según lugar de nacimiento.

6. **`finalizacion_nivel_educativo` (`CH13`):**  
   - Este dato puede ser redundante si ya se está utilizando el nivel educativo máximo alcanzado (`nivel_educativo_maximo`) para medir la relación con los ingresos.

---

## Datos contextuales que podrían ser irrelevantes

- **`año` y `trimestre`:**  
   - Si no se planea realizar un análisis temporal o identificar tendencias en ingresos o empleo a lo largo del tiempo.
