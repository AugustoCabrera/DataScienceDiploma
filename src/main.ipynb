## Elementos de un Diagrama de Caja


  <td align="center" style="padding: 0 20px;">
        <img src="../img/image4.png" alt="" width="400">
      </td>

Un diagrama de caja (o *boxplot*) resume un conjunto de datos num√©ricos mediante cinco valores principales y posibles valores at√≠picos. A continuaci√≥n, se describen sus elementos:

### 1. **Caja (o rect√°ngulo)**
- **L√≠mite inferior de la caja:** Representa el primer cuartil (**Q1**), que delimita el 25% inferior de los datos.
- **L√≠mite superior de la caja:** Representa el tercer cuartil (**Q3**), que delimita el 75% inferior de los datos (o el 25% superior).
- **L√≠nea dentro de la caja:** Muestra la mediana (**Q2**), que es el valor central del conjunto de datos.

### 2. **Bigotes**
- Los bigotes son l√≠neas que se extienden desde los l√≠mites de la caja hasta los valores m√°s extremos que no se consideran at√≠picos.
- Se calculan generalmente como:
  - **L√≠mite inferior:** $Q1 - 1.5 \times \text{IQR}$
  - **L√≠mite superior:** $Q3 + 1.5 \times \text{IQR}$
  - Donde $ \text{IQR} $ (rango intercuart√≠lico) es $ Q3 - Q1 $.

### 3. **Valores at√≠picos (outliers)**
- Son puntos individuales fuera de los l√≠mites definidos por los bigotes.
- Se representan con s√≠mbolos como c√≠rculos o asteriscos.

### 4. **Rango intercuart√≠lico (IQR)**
- Aunque no se dibuja directamente, es una medida importante para calcular los bigotes:
  $$
  \text{IQR} = Q3 - Q1
  $$

### 5. **Escala num√©rica**
- El eje del diagrama, generalmente el eje x o y, muestra los valores del conjunto de datos.

Este diagrama es √∫til para identificar la **dispersi√≥n**, **simetr√≠a** y **valores at√≠picos** de los datos.

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['ingreso_total'], color='lightgreen')
plt.title('Boxplot de Ingresos Totales')
plt.xlabel('Ingreso Total')
plt.show()
```

El boxplot que muestras parece indicar que los datos de ingreso total est√°n altamente concentrados en valores bajos, mientras que hay un n√∫mero significativo de valores at√≠picos (outliers) en rangos altos. Esto es com√∫n en distribuciones de ingresos, ya que la mayor√≠a de las personas tienen ingresos dentro de un rango m√°s estrecho, y unos pocos casos extremos (como altos ejecutivos o empresarios) tienen ingresos muy elevados.

## Mejora del An√°lisis del Gr√°fico

Para entender mejor la distribuci√≥n:

```python
# Calcular estad√≠sticas descriptivas
Q1 = df['ingreso_total'].quantile(0.25)  # Primer cuartil
Q2 = df['ingreso_total'].median()        # Mediana
Q3 = df['ingreso_total'].quantile(0.75)  # Tercer cuartil
IQR = Q3 - Q1                                 # Rango intercuart√≠lico
upper_bound = Q3 + 1.5 * IQR                  # L√≠mite superior
lower_bound = Q1 - 1.5 * IQR                  # L√≠mite inferior

# Valores superiores al l√≠mite superior
outliers = df[df['ingreso_total'] > upper_bound]
num_outliers = len(outliers)

# Mostrar estad√≠sticas
print("Primer cuartil (Q1):", Q1)
print("Mediana (Q2):", Q2)
print("Tercer cuartil (Q3):", Q3)
print("Rango intercuart√≠lico (IQR):", IQR)
print("L√≠mite superior:", upper_bound)
print("N√∫mero de valores superiores al l√≠mite superior:", num_outliers)

# Crear el boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['ingreso_total'], color='lightgreen')
plt.title('Boxplot de Ingresos Totales con L√≠mites')
plt.xlabel('Ingreso Total')

# A√±adir l√≠neas para los cuartiles y la mediana
plt.axvline(Q1, color='blue', linestyle='--', label='Q1 (25%)')
plt.axvline(Q2, color='red', linestyle='-', label='Mediana (50%)')
plt.axvline(Q3, color='blue', linestyle='--', label='Q3 (75%)')

# A√±adir l√≠mites al gr√°fico
plt.axvline(upper_bound, color='orange', linestyle='--', label='L√≠mite Superior')
plt.axvline(lower_bound, color='orange', linestyle='--', label='L√≠mite Inferior')

# Mostrar la leyenda
plt.legend()
plt.show()
```

```python
# Valores dentro de la caja (entre Q1 y Q3)
within_box = df[(df['ingreso_total'] >= Q1) & (df['ingreso_total'] <= Q3)]
num_within_box = len(within_box)

# Valores fuera de los bigotes
outliers = df[(df['ingreso_total'] < lower_bound) | (df['ingreso_total'] > upper_bound)]
num_outliers = len(outliers)

# Total de datos
total_data = len(df)

# Porcentajes
percent_within_box = (num_within_box / total_data) * 100
percent_outliers = (num_outliers / total_data) * 100

# Mostrar resultados
print("Porcentaje de datos dentro de la caja (Q1-Q3): {:.2f}%".format(percent_within_box))
print("Porcentaje de datos fuera de los bigotes: {:.2f}%".format(percent_outliers))
```

```python
# Definir el salario m√≠nimo
salario_minimo = 156000

# Filtrar los datos
superan_salario_minimo = df[df['ingreso_total'] > salario_minimo]
no_superan_salario_minimo = df[df['ingreso_total'] <= salario_minimo]

# Contar los datos
num_superan = len(superan_salario_minimo)
num_no_superan = len(no_superan_salario_minimo)

# Total de datos
total_datos = len(df)

# Calcular porcentajes
percent_superan = (num_superan / total_datos) * 100
percent_no_superan = (num_no_superan / total_datos) * 100

# Mostrar resultados
print("N√∫mero de ingresos que superan el salario m√≠nimo:", num_superan)
print("Porcentaje de ingresos que superan el salario m√≠nimo: {:.2f}%".format(percent_superan))
print("N√∫mero de ingresos que no superan el salario m√≠nimo:", num_no_superan)
print("Porcentaje de ingresos que no superan el salario m√≠nimo: {:.2f}%".format(percent_no_superan))
```

De un total de **16,846 ingresos analizados**, se observa lo siguiente:

1. **El 62.69% de los ingresos (10,561 casos)** superan el salario m√≠nimo de **156,000 pesos**.  
   Esto indica que la mayor√≠a de las personas tienen ingresos por encima de este umbral, lo que podr√≠a reflejar una distribuci√≥n de ingresos relativamente favorable para este grupo.

# Datos del Dataset que Podr√≠an No Ser de Inter√©s

Para determinar qu√© datos podr√≠an no ser de inter√©s, evaluamos si cumplen con alguna de las siguientes caracter√≠sticas:

1. **Redundancia:** Datos repetidos o que ya est√°n representados en otras columnas.
2. **Relevancia:** Datos que no aportan valor al an√°lisis actual o que no est√°n relacionados con los objetivos del estudio.
3. **Falta de variabilidad:** Columnas donde todos los valores (o casi todos) son iguales, lo que no permite identificar diferencias significativas.

---

## Posibles datos no relevantes para ciertos an√°lisis

1. **`peso`:**  
   - Representa el factor de ponderaci√≥n de cada observaci√≥n. Si no se est√° realizando un an√°lisis representativo a nivel poblacional, podr√≠a no ser √∫til.

2. **`habilidad_lectura_escritura`:**  
   - Tiene poco valor en un an√°lisis econ√≥mico, ya que la mayor√≠a de los individuos empleados probablemente tengan esta habilidad, y podr√≠a no mostrar variabilidad significativa.

3. **`asistencia_educativa` (`CH10`):**  
   - Indica si el individuo asiste a un establecimiento educativo. Esto podr√≠a ser irrelevante si el an√°lisis est√° enfocado en ingresos, empleo o regi√≥n.

4. **`tipo_institucion_educativa` (`CH11`):**  
   - Si el inter√©s principal est√° en los ingresos o el empleo, este dato podr√≠a no ser relevante, ya que se relaciona m√°s con la educaci√≥n de ni√±os o j√≥venes.

5. **`nacionalidad`:**  
   - Podr√≠a ser irrelevante si no se planea investigar diferencias en ingresos o empleo seg√∫n lugar de nacimiento.

6. **`finalizacion_nivel_educativo` (`CH13`):**  
   - Este dato puede ser redundante si ya se est√° utilizando el nivel educativo m√°ximo alcanzado (`nivel_educativo_maximo`) para medir la relaci√≥n con los ingresos.

---

## Datos contextuales que podr√≠an ser irrelevantes

- **`a√±o` y `trimestre`:**  
   - Si no se planea realizar un an√°lisis temporal o identificar tendencias en ingresos o empleo a lo largo del tiempo.
